```python
# Mengimport library pandas untuk membaca dataset
# Mengimport matplotlib.pyplot membuat dan mengontrol berbagai jenis plot
# Mengimport seaborn untuk visualisasi data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('/content/epl_player_stats_24_25.csv', delimiter=";", encoding='latin-1')


# Set visualization style
sns.set_style("whitegrid")
%matplotlib inline
```

# 3. Mengecek Data yang Hilang (Missing Value)

```python
# Mengecek data yang missing
df.isna().sum()
```

```python
# Menampilkan nilai statistik

df.describe().T
```

```python
# cek statiscial summary pada suatu value kolum
df['Appearances'].describe()
```

```python
 #Mengatasi missing value
for column in df.columns:
    if df[column].dtype == 'object':
        # Jika kolom bertipe object, isi dengan mode
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        # Jika kolom bertipe numerik, isi dengan mean
        df[column].fillna(df[column].mean(), inplace=True)
```

```python
# Mengecek apakah ada duplicate di seluruh kolom
check_duplicate = df.duplicated().sum()

print(f"Jumlah data yang duplikat = {check_duplicate}")
```

# Top 10 player Goal Score Ammount

```python
print(df.columns.tolist())


top_scorers = df.nlargest(10, 'Goals')[['Player Name', 'Club', 'Position' ,'Goals']]
top_scorers = top_scorers.reset_index(drop=True)

top_scorers
```

```python
# Menampilkan visual chart bar

top_scorer = (
    df[['Player Name', 'Goals']]
    .sort_values(by='Goals', ascending=False)
    .head(10)
)

plt.figure(figsize=(12, 6))
ax = sns.barplot(
    data=top_scorer,
    x='Player Name',
    y='Goals',
    hue='Player Name',
    palette='viridis',
    dodge=False,
    width=0.6
)

plt.title('Top 10 Goal Scorers â€“ 2024/25 Season', fontsize=14)
plt.xlabel('Player', fontsize=12)
plt.ylabel('Goals', fontsize=12)
plt.xticks(rotation=30, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.legend().remove()
plt.tight_layout()
plt.show()
```

# TOP Assist & Appearances

```python
import pandas as pd
from IPython.display import display, HTML

print(df.columns.tolist())

top_assists = df.nlargest(10, 'Assists')[['Player Name', 'Club', 'Assists']]
top_assists = top_assists.reset_index(drop=True)

# Filter data untuk mengecualikan kiper
df_non_gk = df[df['Position'] != 'GKP'].copy()

# Hitung top 10 pemain non-kiper dengan menit bermain terbanyak
top_appearances = df_non_gk.nlargest(10, 'Minutes')[['Player Name', 'Club', 'Appearances', 'Minutes']]
top_appearances = top_appearances.reset_index(drop=True)

# Gabungkan representasi HTML dari kedua tabel
html_output = f"""
<div style="display: flex;">
    <div style="margin-right: 20px;">
        <h3>Top 10 Players by Assists:</h3>
        {top_assists.to_html()}
    </div>
    <div>
        <h3>Top 10 Players by Appearances (Excluding Goalkeepers):</h3>
        {top_appearances.to_html()}
    </div>
</div>
"""

# Tampilkan HTML
display(HTML(html_output))
```

# KORELASI BETWEEN SAVE & XGOT CONCEDED

```python
#Membuat scatter plot
plt.figure(figsize=(6, 6))
sns.scatterplot(data=df, x='Saves', y= "xGoT Conceded", color='blue')
plt.title('Save vs. xGoT Conceded')
plt.xlabel('Saves')
plt.ylabel('xGoT Conceded')
plt.show()

#korelasi plot
corr_val = df['Saves'].corr(df['xGoT Conceded'])
print(f"Pearson correlation between Saves and  'xGoT Conceded: {corr_val:.2f}")
```
